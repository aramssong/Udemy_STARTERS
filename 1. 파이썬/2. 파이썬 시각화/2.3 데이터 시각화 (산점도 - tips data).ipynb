## 4.1 지불 금액 - 팁의 관계

```python
plt.scatter(tips['total_bill'], tips['tip']
           , s = tips['size'] * 50   # 점의 크기 적용
           , alpha = 0.5)            # 점의 투명도

plt.title('지불금액과 팁의 관계')

plt.xlabel('지불 금액(달러)')
plt.ylabel('팁(달러)')

plt.show()
```

## 4.2 지불 금액 - 팁의 관계 (+ 성별 표시)

```python
# 성별에 따른 color 컬럼 추가

def set_color(x):
    if x == 'Male':
        return 'blue'
    elif x == 'Female':
        return 'red'

tips['color'] = tips['sex'].apply(set_color)
tips.head()
```

```python
plt.scatter(tips['total_bill'], tips['tip']
           , s = tips['size'] * 50
           , alpha = 0.5
           , c = tips['color'])

plt.title('지불금액과 팁의 관계')

plt.xlabel('지불 금액(달러)')
plt.ylabel('팁(달러)')

plt.show()
```
